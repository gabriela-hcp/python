# **Análise de Dados com Aprendizado de Máquina (Machine Learning) com Python e Pandas**

O aprendizado/aprendizagem automático(a)/de máquina é um subcampo da Engenharia e da Ciência da Computação que evoluiu do estudo de reconhecimento de padrões e da teoria do aprendizado computacional em inteligência artificial.

É uma tecnologia onde os computadores tem a capacidade de aprender de acordo com as respostas esperadas por meio associações de diferentes dados, os quais podem ser imagens, números e tudo que essa tecnologia possa identificar. 

Os algorítmos de ML são divididos em **Supervisionado**, **Não Supervisionado** e por **Reforço**.

image.png

## **ML Supervisionado**

O aprendizado supervisionado geralmente começa com um conjunto estabelecido de dados e um certo entendimento de como esses dados são classificados. O aprendizado supervisionado destina-se a encontrar padrões em dados que possam ser aplicados em um processo analítico. Esses dados rotularam recursos que definem o significado dos dados.

Por exemplo, é possível criar um aplicativo de machine learning que distinga entre milhões de animais, com base em imagens e descrições escritas.

O supervisionado se divide em **Classificação** e **Regressão**.

## **ML Supervisionado de Classificação**

**Classificação**: baseia-se em prever a categoria de uma observação dada. Aqui, procura-se estimar um “classificador” que gere como saída a classificação qualitativa de um dado não observado com base em dados de entrada (que abrangem observações com classificações já definidas).

**Objetivo**: prever variável categórica.

**Métricas**: Acurácia, Precisão, Recall, ROC e AUC.

**Exemplo**: um classificador que utilize dados não observados de um paciente e classifique-o como doente ou não-doente.

***Vamos ver os passos para trabalharmos com algorítmos de aprendizado de máquina supervisionado para classificação.***

### **Identificar o Problema**

O primeiro passo é identificar o seu problema.

Seu problema pode ser baseado em bases famosas, prontas para estudar ou a partir de um sistema conhecido, onde possa ser exportada os dados.

### **Obter os dados**

Esse problema pode já existir um conjunto de dados (Dataset) pronto, já existente.

Pode ser buscado em dados aberto e/ou públicos de sites já citados.

Em outros casos, o conjunto de dados (Dataset) pode ser obtido através de resultados de consultas SQL sobre a base de dados do seu sistema, por exemplo.

Também podemos capturar os dados de várias fontes e após cruzar, gerando um novo conjunto de dados (Dataset) para ser analisado.

### **Separar em Treino e Teste**



É importante separar os dados em Treino e Teste ou Treino, Validação e Teste, antes de iniciar as manipulações dos dados, a limpeza dos dados (data cleaning).

Após separar é importante executar todos as tratativas em todas as bases.

Essa separação, logo no início, se faz necessária para evitar vieses.

Este método de dividir o conjunto de dados (Dataset) é chamado de **holdout**.
